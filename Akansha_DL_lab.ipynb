AND Gate

```python
import numpy as np

def pitts_neuron(x, w, theta):
    """McCulloch-Pitts neuron"""
    return 1 if np.dot(x, w) >= theta else 0

print("AND Gate")
w = np.array([1, 1])
theta = 2
for x in [(0,0), (0,1), (1,0), (1,1)]:
    print(pitts_neuron(np.array(x), w, theta))
```

Artificial Neuron


```python
import numpy as np

def art_neuron(x, w, b, activation='step'):
    z = np.dot(x, w) + b
    if activation == 'step':
        return 1 if z >= 0 else 0
    elif activation == 'sigmoid':
        return 1 / (1 + np.exp(-z))
    elif activation == 'relu':
        return max(0, z)

x = np.array([1, 2, 3])
w = np.array([0.1, 0.2, 0.3])
b = 0.1
print('step output:', art_neuron(x, w, b, activation='step'))
print('sigmoid output:', art_neuron(x, w, b, activation='sigmoid'))
print('relu output:', art_neuron(x, w, b, activation='relu'))
```
